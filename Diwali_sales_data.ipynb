## EXPLORATORY DATA ANALYSIS

##### GENDER

```python
df.columns
```

```python
#to plot gender data in bar
sns.countplot(x="Gender",data=df)
```

```python
# to initialize bar value
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#group values according to the gender and sum of amuont in descinding order
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#to see the above data in bars
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

##### From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

##### AGE

```python
df.columns
```

```python
# different age group count
sns.countplot(data=df,x="Age Group")
```

```python
#Age group according to gender
sns.countplot(x="Age Group",hue="Gender",data=df)
```

```python
# same graph with values
ax=sns.countplot(x="Age Group",hue="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
colorBar=["r","g","violet","b","k","hotpink","darkred"]
sns.barplot(x="Age Group",y="Amount",data=sales_age,hue='Amount',legend=False,palette=colorBar) # palette assign the color for each bar
```

##### From the above graphs we can see that most of tye buyers are of age group between 26-35 years females

##### STATE

```python
# total numbers of order fromtop 10 states
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
colorBar=["k","g","violet","b","r","hotpink","darkred","#4caf50","blue","green"]
sns.set(rc={"figure.figsize":(17,5)}) # it solvees the problem of x labels name (width,height)
sns.barplot(x="State",y="Orders",data=sales_state,hue="Orders",legend=False,palette=colorBar)
```

```python
#total amount/sales from top 10 state
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
colorBar=["k","g","violet","b","r","hotpink","darkred","#4caf50","blue","green"]
sns.set(rc={"figure.figsize":(17,5)})
sns.barplot(x="State",y="Amount",data=sales_state,hue="Amount",legend=False,palette=colorBar)
```

##### From the above graph we can see that most of the buyers are from Uttarpradesh,Maharastra and Karnataka

##### MARITAL STATUS

```python
df.columns
```

```python
# to see count of married and unmarried buyers
ax=sns.countplot(x="Shaadi_Status",data=df)
sns.set(rc={"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married=df.groupby(["Shaadi_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(7,5)})
sns.barplot(x="Shaadi_Status",y="Amount",data=sales_married,hue="Gender")
```

##### From the above graph we can see that most of the buyers are married(women) and they have high purchasing power

##### OCCUPATION

```python
## to see count of occupation
sns.set(rc={"figure.figsize":(20,5)})
colorBar=["k","g","violet","b","r","hotpink","darkred","#4caf50","blue","green","pink","r","hotpink","darkred","#4caf50"]
ax=sns.countplot(x="Occupation",data=df,palette=colorBar,hue="Occupation",legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Occupation",y="Amount",data=sales_occ,hue="Occupation",legend=False)
```

##### From the above graph we can see that most of the buyers are wroking at IT,Healthcare Sector and Aviation 

##### PRIDUCT CATAGORY

```python
df.columns
```

```python
## to see count of product catagory
sns.set(rc={"figure.figsize":(25,5)})
colorBar=["k","g","violet","b","r","hotpink","darkred","#4caf50","blue","green","pink","r","hotpink","darkred","#4caf50","green","pink","darkred"]
ax=sns.countplot(x="Product_Category",data=df,palette=colorBar,hue="Product_Category",legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Product_Category",y="Amount",data=sales_occ,hue="Product_Category",legend=False)
```

##### From the above graph we can see that most of te sold products are from Food ,footwear and electronics catagory

##### PRODUCT ID

```python
# top 10 most sold products
sales_occ=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(x="Product_ID",y="Orders",data=sales_occ,hue="Product_ID",legend=False)
```

```python
# top 10 most sold products (same as above graph)
fig1,ax1=plt.subplots(figsize=(12,7))
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
```
